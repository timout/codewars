### Create a funnel

Create a Funnel Data Structure with max capacity: 15. It consists of three basic methods: fill(), drip() and toString()/to_s/__str__.

```python

def l_up(cur, level): return cur + level + 1
def r_up(cur, level): return cur + level + 2

class Funnel:
    def __init__(self):
        self.size = 15
        self.levels = 5
        self.fun = [None] * self.size
        
    def fill(self, *values):
        if len(values) == 0: return
        i,v = 0,0
        while i < self.size and v < len(values):
            if self.fun[i] is None:
                self.fun[i] = values[v]
                v += 1
            i += 1
        
    def drip(self):
        if self.fun[0] is None: return None
        r = self.fun[0]
        cur = 0
        level = 0
        while level < self.levels:
            self.fun[cur] = None
            l_up_pos = l_up(cur, level)
            r_up_pos = r_up(cur, level)
            level += 1
            left = self.get_weight(l_up_pos, level)
            right = self.get_weight(r_up_pos, level)
            #print('left_w {}:{} right_w {}:{}'.format(l_up_pos, left, r_up_pos,right))
            if left > right or (left == right and left > 0):
                self.fun[cur] = self.fun[l_up_pos]
                cur = l_up_pos
            elif right > left:
                self.fun[cur] = self.fun[r_up_pos]
                cur = r_up_pos
            else:
                level = self.levels
        return r
    
    #number of all not None items over element + 1(element) or 0 if element is None
    def get_weight(self, i, level):
        if level >= self.levels or self.fun[i] is None: return 0
        w = 1
        l = 2
        while level < self.levels:
            i = l_up(i, level)
            w += sum([ 0 if j is None else 1 for j in self.fun[i:i+l] ])
            level = level + 1
            l += 1
        return w
            
        
    def __str__(self):
        s = self.size
        res = []
        for i in range(self.levels, 0, -1):
            s = s - i
            l = (" " * (self.levels-i)) + "\\" + " ".join(map(lambda v: " " if v is None else str(v), self.fun[s:s+i])) + "/"
            res.append(l)
        return "\n".join(res)
            
        
```

```python
fun = Funnel()
#print(fun)
fun.fill(0,1,2,3,4)
#print(fun)
fun.fill(0,1,2,3,4,5,6,7,6,7)
d = fun.drip()
while not (d is None):
    print(d)
    print(fun)
    d = fun.drip()
```
